```python
import re
import string



def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.strip()
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
data['Questions'].apply(clean_text)
questions = list(data['Questions'].apply(clean_text))
answers   = list(data['Answers'].apply(clean_text))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(questions)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def faq_chatbot(user_input):
    user_vec = vectorizer.transform([user_input])
    scores   = cosine_similarity(user_vec, x)[0]
    best_idx = np.argmax(scores)
    
    if scores[best_idx] > 0.2:  # threshold
        return answers[best_idx]
    else:
        return "Sorry, I don't have an answer for that."
```

```python
print(faq_chatbot('What does it mean to have a mental illness?'))
```
